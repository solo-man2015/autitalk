```python
# Create architect to provide the architecture for Autitalk
from langchain_google_genai import Agent
from langchain_google_genai import Task

architect = Agent(
  role='Senior Application architect',
  goal='You are creating an architecture for a mobile/tablet based learning app with easy to use GUI that allows autistic children to talk to it, understand them as best as possible, provide visual options to them on what they actually require, then speak out the correct sentence, it can also prompt the child to say the sentense aloud or if it is too difficult, then with a tap of a button speak out the correct expression or need, like "Mommy, I want to have some cereal now"',
  backstory='You are an expert application architect, who can provide architecture for complex software problems with AI on any platform. ',

  allow_code_execution=True  # This lets the agent actually "test" the code
)

# Create a task for the Architect
coding_task = Task(
  description='You are creating an architecture for a mobile/tablet based app with easy to use GUI that allows autistic children to talk to it, understand them as best as possible, provide visual options to them on what they actually require, then speak out the correct sentense, it can also prompt the child to say the sentense aloud or if it is too difficult, then with a tap of a button speak out the correct expression or need, like "Mommy, I want to have some cereal now", provide the necessary, architecture, software components, tools required for development',
  expected_output='Complete document on Overall architecture, tools, software modules, workflows, test execution, deployment mechanism for Android, iOS, how to run it on simulators for iOS on mac for testing',
  agent=architect
)
```
